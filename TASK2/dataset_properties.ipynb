```python
from datasets import Dataset, DatasetDict, load_dataset
import pandas as pd
import matplotlib as plt
```

```python
dataset = load_dataset("DeveloperOats/DBPedia_Classes")
```

```python
train_df = pd.DataFrame(dataset["train"])
test_df = pd.DataFrame(dataset["test"])
```

```python
# Plotting the l1 label distribution

train_df['l1'].value_counts().plot(kind='barh', title="l1 labels")
file_name = "_DIST"
plt.pyplot.savefig('./figures/'+file_name)

#plt.pyplot.show()

```

```python
# Plotting l2 label distribution for every l1 label

for label1 in set(train_df.loc[:, 'l1']):
    graph = train_df[train_df['l1'] == label1].loc[:, 'l2'].value_counts().plot(kind='barh', title=label1, color='red')
    file_name = label1+"_DIST"
    plt.pyplot.savefig('./figures/'+file_name)
    #plt.pyplot.show()
```

```python
# Plotting l3 label distribution for every l2 label (for every l1 label)

for label1 in set(train_df.loc[:, 'l1']):
    for label2 in set(train_df[train_df['l1']==label1].loc[:, 'l2']):
        title_str=label1+">"+label2    
        graph2 = train_df[(train_df['l1']==label1) & (train_df['l2']==label2)]['l3'].value_counts().plot(kind='barh', title=title_str, color='green')
        file_name2=label1+"-"+label2+"_DIST"
        plt.pyplot.savefig('./figures/'+file_name2)
        #plt.pyplot.show() 
```
